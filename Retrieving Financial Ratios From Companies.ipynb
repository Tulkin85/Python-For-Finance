```python
def financialratios(company):
    fr = requests.get(f'https://financialmodelingprep.com/api/v3/ratios/{company}?limit=40&apikey={demo}')
    fr = fr.json()
    current_fr = fr[0]
    df = pd.DataFrame(current_fr.items(), columns = ['Ratio', company])
    df = df.iloc[2:]
    return df
```

```python
tickers = ['VST', 'PEG', 'EXC']

x = financialratios('NRG')

for ticker in tickers:
    y = financialratios(ticker)
    x = x.merge(y, on='Ratio')
```
