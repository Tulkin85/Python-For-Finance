```python
def balancesheet(company):
    r = requests.get(f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{company}?period=quarter&limit=400&apikey={demo}')
    bs = r.json()
    bs = pd.DataFrame.from_dict(bs)
    bs = bs.T
    bs.columns = bs.iloc[0]
    bs = bs.iloc[5:-2, :6]
    cols = bs.columns
    bs[cols] = bs[cols].apply(pd.to_numeric, errors = 'coerce')
    bs = bs.iloc[0:17, :]
    assetsq1 = bs.iloc[16,0]
    assetsq2 = bs.iloc[16,1]
    assetsq3 = bs.iloc[16,2]
    assetsq4 = bs.iloc[16,3]
    assetsq5 = bs.iloc[16,4]
    assetsq6 = bs.iloc[16,5]
    allassets = [assetsq1,assetsq2,assetsq3,assetsq4,assetsq5,assetsq6 ]
    bs[cols] = bs[cols] / allassets
    pd.options.display.float_format = '{:.2f}%'.format
    bs = bs * 100
    return bs
    
```
