```python
import pandas as pd
import requests
import json
import plotly.graph_objects as go
```

```python
url = 'https://api.db.nomics.world/v22/series/WB/WDI/FP.CPI.TOTL.ZG-XU?observations=1'
```

```python
def checkindicator(url):
    r = requests.get(url)
    r = r.json()
    
    periods = r['series']['docs'][0]['period']
    values = r['series']['docs'][0]['value']
    dataset = r['series']['docs'][0]['dataset_name']
    
    indicators = pd.DataFrame(values, index = periods)
    indicators.columns = [dataset]
    return indicators
```

```python
inflation = checkindicator(url)
inflation
```

```python
bond_yields = checkindicator('https://api.db.nomics.world/v22/series/Eurostat/ei_mfir_m/M.NSA.NAP.MF-LTGBY-RT.AT?observations=1')
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x = bond_yields.index,
    y = bond_yields['Interest rates - monthly data'],
    name = 'Bond Yields',
    line_color = 'deepskyblue'
))

fig.add_trace(go.Scatter(
    x = inflation.index,
    y = inflation['World Development Indicators'],
    name = 'Inflation',
    line_color = 'black'
))

fig.update_layout(xaxis_range=['2003-01-01', '2019-12-31'],
                 title_text = 'Inflation',
                 xaxis_rangeslider_visible = True)

fig.show()
```
